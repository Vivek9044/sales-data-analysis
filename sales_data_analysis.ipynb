```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv(r"C:\Users\kasau\Downloads\archive\sales_data_sample.csv", encoding="latin1")

monthly_sales = data.groupby("MONTH_ID")["SALES"].sum()

monthly_sales.plot(kind="bar")

plt.title("Monthly Sales Analysis")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.show()
```

```python
top_products = data.groupby("PRODUCTLINE")["SALES"].sum().sort_values(ascending=False)

top_products.plot(kind="bar")

plt.title("Sales by Product Line")
plt.xlabel("Product Line")
plt.ylabel("Sales")

plt.show()
```

```python
country_sales = data.groupby("COUNTRY")["SALES"].sum().sort_values(ascending=False).head(10)

country_sales.plot(kind="bar")

plt.title("Top Countries by Sales")
plt.xlabel("Country")
plt.ylabel("Sales")

plt.show()
```
